El objetivo es identificar la especie de la cual proviene el canto, mediante el analisis de fourier

```python
#t,f,Sxx = tyn.spectrograma2D(canal1,sample_rate, 0,7500 , log = 0, a = 8 , b = 6, cmapp= "plasma")
#t1,f1,Sxx1 = tyn.spectrograma2D(canal11,sample_rate1, 0,7500 , log = 0, a = 7 , b = 5, cmapp= "plasma")
t,f,Sxx = tyn.spectrograma2D(samples,sample_rate, 0,7500 , log = 0, a = 8 , b = 6, cmapp= "plasma")
t1,f1,Sxx1 = tyn.spectrograma2D(samples1,sample_rate1, 0,7500 , log = 0, a = 8 , b = 6, cmapp= "plasma")
```

```python
h, p = tyn.huella_p(f,t,Sxx,"Brotogeris jugularis.",0.2)
```

```python
h1, p1 = tyn.huella_p(f1,t1,Sxx1,"Forpus",0.2)
```

```python
plt.figure(figsize= (8,8))
plt.stem(h, p, 'b', 'bo', label = "paco_1")
plt.stem(h1,p1  , 'k', 'ko', label = "paco_2")
plt.ylabel('Potencia', fontsize = 16)
plt.xlabel('Frecuencia [Hz]', fontsize = 16)
plt.legend()
plt.show()
```
